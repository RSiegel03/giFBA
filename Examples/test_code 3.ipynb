```python
import gifba

# load models and media:
sim = "test"
models, media = gifba.utils.load_simple_models(sim) # sim. 4a

# initialize community
community = gifba.gifbaObject(models, media, rel_abund=[0.2, 0.8], step_size=0.01)
# run iterations
media_flux, org_flux = community.run_gifba(iters=1000, method="pfba", early_stop=False, v=True)
```

```python
media_flux
```
