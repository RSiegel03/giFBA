# Messing with Convergence tradeoff
Slows time to convergence, but will increase the iteration interaction


```python
import gifba
import cobra as cb

m1 = cb.io.load_matlab_model('AGORA2_Models/Bifidobacterium_longum_infantis_ATCC_15697.mat')
m2 = cb.io.load_matlab_model('AGORA2_Models/Eubacterium_hallii_DSM_3353.mat')

# AGORA Format
media = {
    'EX_o2(e)': 0, #aerobic/anaerobic
    'EX_h2o(e)': -1000,
    'EX_pi(e)': -1000,
    'EX_fe2(e)': -1000,
    'EX_fe3(e)': -1000,
    'EX_zn2(e)': -1000,
    'EX_so4(e)': -1000,
    'EX_cu2(e)': -1000,
    'EX_k(e)': -1000,
    'EX_mg2(e)': -1000,
    'EX_mn2(e)': -1000,
    'EX_cd2(e)': -1000,
    'EX_cl(e)': -1000,
    'EX_ca2(e)': -1000,
    'EX_cobalt2(e)': -1000,
    'EX_glc_D(e)': -10,
    'EX_nh4(e)': -20,

    'EX_ribflv(e)': -1000,
    'EX_pnto_R(e)': -1000,
    'EX_nac(e)': -1000,
    'EX_his_L(e)': -1000,
    'EX_asn_L(e)': -1000,
    'EX_glycys(e)': -1000,

    'EX_lys_L(e)': -1000,
    'EX_ala_L(e)': -1000,
    'EX_met_L(e)': -1000,
    'EX_leu_L(e)': -1000,
    'EX_hxan(e)': -1000,    
    'EX_ser_L(e)': -1,
}
```

```python
# initialize community
community_1 = gifba.gifbaObject([m1, m2], media, rel_abund=[0.5, 0.5], step_size=1)
community_075 = gifba.gifbaObject([m1, m2], media, rel_abund=[0.5, 0.5], step_size=0.75)
# community_05 = gifba.gifbaObject([m1, m2], media, rel_abund=[0.5, 0.5], step_size=0.5)
# community_025 = gifba.gifbaObject([m1, m2], media, rel_abund=[0.5, 0.5], step_size=0.25)
# community_01 = gifba.gifbaObject([m1, m2], media, rel_abund=[0.5, 0.5], step_size=0.1)
# community_001 = gifba.gifbaObject([m1, m2], media, rel_abund=[0.5, 0.5], step_size=0.01)
# run iterations
its = 1000
media_flux_1, org_flux_1 = community_1.run_gifba(iters=its, method="pfba", early_stop=True, v=False)
media_flux_075, org_flux_075 = community_075.run_gifba(iters=its, method="pfba", early_stop=True, v=False)
# media_flux_05, org_flux_05 = community_05.run_gifba(iters=its, method="pfba", early_stop=True, v=False)
# media_flux_025, org_flux_025 = community_025.run_gifba(iters=its, method="pfba", early_stop=True, v=False)
# media_flux_01, org_flux_01 = community_01.run_gifba(iters=its, method="pfba", early_stop=True, v=False)
# media_flux_001, org_flux_001 = community_001.run_gifba(iters=its, method="pfba", early_stop=True, v=True)
```

```python
[community_1.iter_converged, community_075.iter_converged, 
community_05.iter_converged, community_025.iter_converged, 
community_01.iter_converged, community_001.iter_converged]
```

```python
import matplotlib.pyplot as plt
import numpy as np
# with community_001 as comm:
# 	plt.plot(comm.env_fluxes["EX_but(e)"], label='butyrate')
# 	plt.plot(comm.env_fluxes["EX_glc_D(e)"], label='glucose')
# 	plt.plot(comm.env_fluxes["EX_lac_L(e)"], label='lactate')
# 	plt.plot(comm.env_fluxes["EX_ac(e)"], label='acetate')
# 	plt.legend()

fig, ax = plt.subplots(5, 6, sharey='row', figsize=(20, 15))
for idx, (step_size, community) in enumerate(zip(
	[1, 0.75, 0.5, 0.25, 0.1, 0.01],
	[community_1, community_075, community_05, community_025, community_01, community_001]
)):
	model_idx = 1
	ax[0, idx].plot(community.org_fluxes.loc[model_idx, "EX_but(e)"], label='butyrate')
	ax[0, idx].set_xlim(-0.1 * community.iter_converged, community.iter_converged * 1.5)
	
	ax[1, idx].plot(community.org_fluxes.loc[model_idx, "EX_glc_D(e)"], label='glucose')
	ax[1, idx].set_xlim(-0.1 * community.iter_converged, community.iter_converged * 1.5)

	ax[2, idx].plot(community.org_fluxes.loc[model_idx, "EX_lac_L(e)"], label='lactate')
	ax[2, idx].set_xlim(-0.1 * community.iter_converged, community.iter_converged * 1.5)
	
	ax[3, idx].plot(community.org_fluxes.loc[model_idx, "EX_ac(e)"], label='acetate')
	ax[3, idx].set_xlim(-0.1 * community.iter_converged, community.iter_converged * 1.5)

	ax[4, idx].plot(community.org_fluxes.loc[model_idx, "EX_biomass(e)"], label='biomass')
	ax[4, idx].set_xlim(-0.1 * community.iter_converged, community.iter_converged * 1.5)
	
	ax[0, idx].set_title(f'Step Size: {step_size}\n {community.iter_converged} iters')
	if idx == 0:
		ax[0, idx].legend()
		ax[1, idx].legend()
		ax[2, idx].legend()
		ax[3, idx].legend()
		ax[4, idx].legend()
```
