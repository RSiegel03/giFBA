```python
import gifba
import cobra as cb
```

```python
m1 = cb.io.load_matlab_model('AGORA2_Models/Bifidobacterium_longum_infantis_ATCC_15697.mat')
m2 = cb.io.load_matlab_model('AGORA2_Models/Eubacterium_hallii_DSM_3353.mat')
```

```python
# AGORA Format
media = {
    'EX_o2(e)': 0, #aerobic/anaerobic
    'EX_h2o(e)': -1000,
    'EX_pi(e)': -1000,
    'EX_fe2(e)': -1000,
    'EX_fe3(e)': -1000,
    'EX_zn2(e)': -1000,
    'EX_so4(e)': -1000,
    'EX_cu2(e)': -1000,
    'EX_k(e)': -1000,
    'EX_mg2(e)': -1000,
    'EX_mn2(e)': -1000,
    'EX_cd2(e)': -1000,
    'EX_cl(e)': -1000,
    'EX_ca2(e)': -1000,
    'EX_cobalt2(e)': -1000,
    'EX_glc_D(e)': -10,
    'EX_nh4(e)': -20,

    'EX_ribflv(e)': -1000,
    'EX_pnto_R(e)': -1000,
    'EX_nac(e)': -1000,
    'EX_his_L(e)': -1000,
    'EX_asn_L(e)': -1000,
    'EX_glycys(e)': -1000,

    'EX_lys_L(e)': -1000,
    'EX_ala_L(e)': -1000,
    'EX_met_L(e)': -1000,
    'EX_leu_L(e)': -1000,
    'EX_hxan(e)': -1000,    
    'EX_ser_L(e)': -1,
}
```

```python
# initialize community
community_1_2 = gifba.gifbaObject([m1, m2], media, rel_abund=[0.5, 0.5])
community_1_1 = gifba.gifbaObject([m1, m1], media, rel_abund=[0.5, 0.5])
community_2_2 = gifba.gifbaObject([m2, m2], media, rel_abund=[0.5, 0.5])
# run iterations
media_flux_1_2, org_flux_1_2 = community_1_2.run_gifba(iters=5, method="pfba", early_stop=True, v=True)
media_flux_1_1, org_flux_1_1 = community_1_1.run_gifba(iters=5, method="pfba", early_stop=True, v=True)
media_flux_2_2, org_flux_2_2 = community_2_2.run_gifba(iters=5, method="pfba", early_stop=True, v=True)
```
